```typescript

import { createClient } from 'https://esm.sh/@supabase/supabase-js@2'

// Create a single supabase client for interacting with your database

const supabase = createClient(process.env.SUPABASE_ENTRY_URL, process.env.SUPABASE_PUBLIC_KEY)
```

```typescript

import jsonData from "../resource/courses.json" with { type: "json" };
```

```typescript
for (const course of jsonData.slice(300, 400)) {
  // 관련된 필드를 분리합니다.
  const { address, road_address, contact, operation, ...golfCourseData } = course;
  // 'golf_courses' 테이블에 데이터를 먼저 삽입합니다.
  const { data: golfCourse, error: golfCourseError } = await supabase
    .from('golf_courses')
    .upsert(golfCourseData).select();

  if (golfCourseError) {
    console.error('Error inserting golf course data', golfCourseError);
    continue;
  }
  // id가 안나옴
  console.log(golfCourse[0].name)
  const golfCourseId = golfCourse[0].id;

  // 'addresses' 테이블에 데이터를 삽입합니다.
  if (address) {
    const { error: addressError } = await supabase
      .from('addresses')
      .upsert({ ...address, golf_course_id: golfCourseId });

    if (addressError) {
      console.error('Error inserting address data', addressError);
    }
  }
   // 'road_addresses' 테이블에 데이터를 삽입합니다.
   if (road_address) {
    const { error: roadAddressError } = await supabase
      .from('road_addresses')
      .upsert({ ...road_address, golf_course_id: golfCourseId });

    if (roadAddressError) {
      console.error('Error inserting road address data', roadAddressError);
    }
  }

  // 'contacts' 테이블에 데이터를 삽입합니다.
  if (contact) {
    const { error: contactError } = await supabase
      .from('contacts')
      .upsert({ ...contact, golf_course_id: golfCourseId });

    if (contactError) {
      console.error('Error inserting contact data', contactError);
    }
  }

  // 'operations' 테이블에 데이터를 삽입합니다.
  if (operation) {
    const { error: operationError } = await supabase
      .from('operations')
      .upsert({ ...operation, golf_course_id: golfCourseId });

    if (operationError) {
      console.error('Error inserting operation data', operationError);
    }
  }
}
```
