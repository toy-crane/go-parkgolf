```typescript

import { createClient } from 'https://esm.sh/@supabase/supabase-js@2'

// Create a single supabase client for interacting with your database

const supabase = createClient()
```

```typescript
const { data: golfCourses, error: golfCourseError } = await supabase
    .from('golf_courses')
    .select(`id, name, addresses (address_name), road_addresses (address_name)`)
```

```typescript
const insert_data = golfCourses.map(course => {
  const roadAddress = course.road_addresses.length > 0 ? course.road_addresses[0].address_name : '';
  const address = course.addresses.length > 0 ? course.addresses[0].address_name : '';
  return {
    id: course.id,

    searchable_address: `${address} ${roadAddress} ${course.name}`.trim()
  }
})
```

```typescript

for (const course of insert_data){
  const { data, error } = await supabase
  .from('golf_courses')
  .update({searchable_address: course.searchable_address})
  .eq("id", course.id).select()
}
```
